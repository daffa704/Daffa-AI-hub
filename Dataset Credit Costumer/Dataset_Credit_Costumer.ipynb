```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = pd.read_csv('/content/BankChurners.csv')
```

## Menggunakan dataset

```python
data = {
    'attrition_flag': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'customer_age': [25, 30, 35, 40, 45, 50, 55, 60],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'education_level': ['High School', 'Bachelor', 'Master', 'PhD', 'High School', 'Bachelor', 'Master', 'PhD'],
    'income_category': ['Low', 'Medium', 'High', 'Medium', 'Low', 'High', 'Medium', 'Low']
}
```

```python
df = pd.DataFrame(data)
```

# Visualisasi 1: Pie Chart untuk attrition_flag

Data pelanggan kartu kredit ini memiliki jumlah pelanggan yang berhenti (churn) sama persis dengan jumlah pelanggan yang tetap menggunakan layanan. Artinya, setengah dari pelanggan memilih untuk mengakhiri layanan, dan setengahnya lagi tetap setia.

```python
plt.figure(figsize=(8, 6))
df['attrition_flag'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribusi Attrition Flag')
plt.ylabel('')
plt.show()
```

# Visualisasi 2: Donut Chart untuk gender

Visualisasi ini menunjukkan bahwa dataset "Credit Costumer" memiliki distribusi jenis kelamin yang seimbang, dengan jumlah pelanggan laki-laki dan perempuan yang sama. Hal ini memberikan dasar yang baik untuk analisis dan pemodelan lebih lanjut.

```python
plt.figure(figsize=(8, 6))
plt.pie(df['gender'].value_counts(), labels=df['gender'].value_counts().index, startangle=90, autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Distribusi Gender')
plt.show()
```

# Visualisasi 3: Vertical Bar Chart untuk education_level

Visualisasi menunjukkan jumlah pelanggan "Credit Costumer" berdasarkan tingkat pendidikan (High School, Bachelor, Master, PhD) yang hampir sama rata. Ini berarti tingkat pendidikan tidak banyak memengaruhi jumlah pelanggan. Namun, analisis lebih lanjut perlu dilakukan untuk memahami profil pelanggan secara menyeluruh.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='education_level')
plt.title('Jumlah Pelanggan Berdasarkan Tingkat Pendidikan')
plt.xticks(rotation=45)
plt.show()
```

# Visualisasi 4: Horizontal Bar Chart untuk income_category

Visualisasi menunjukkan bahwa mayoritas pelanggan "Credit Costumer" berada dalam kategori pendapatan "Low", diikuti "Medium", dan "High". Ini mengindikasikan distribusi pendapatan pelanggan yang tidak merata, dengan fokus pada segmen pendapatan rendah

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='income_category')
plt.title('Jumlah Pelanggan Berdasarkan Kategori Pendapatan')
plt.show()
```

# Visualisasi 5: Horizontal Bar Chart untuk customer_age

Visualisasi histogram dengan KDE menunjukkan distribusi usia pelanggan "Credit Costumer" dalam 5 kelompok usia, memberikan gambaran frekuensi dan perkiraan kepadatan probabilitas usia pelanggan.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['customer_age'], bins=5, kde=True)
plt.title('Distribusi Usia Pelanggan')
plt.show()
```

# Visualisasi 6: Countplot untuk memadukan lebih dari satu kolom

Visualisasi diagram batang berkelompok menunjukkan distribusi pelanggan "Credit Costumer" berdasarkan tingkat pendidikan dan kategori pendapatan. Hal ini memungkinkan kita melihat bagaimana pendapatan terdistribusi di setiap tingkat pendidikan.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='education_level', hue='income_category')
plt.title('Jumlah Pelanggan Berdasarkan Tingkat Pendidikan dan Kategori Pendapatan')
plt.xticks(rotation=45)
plt.show()
```

# Visualisasi 7: Scatterplot untuk memadukan lebih dari satu kolom

Visualisasi scatterplot menampilkan hubungan antara usia pelanggan, attrition flag, dan jenis kelamin, memungkinkan identifikasi pola atau tren antar variabel.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='customer_age', y='attrition_flag', hue='gender')
plt.title('Scatterplot Usia Pelanggan dan Attrition Flag')
plt.show()
```

# Visualisasi 8: Heatmap untuk memadukan lebih dari satu kolom

Heatmap menunjukkan korelasi lemah antara usia pelanggan dan kategori pendapatan, mengindikasikan tidak ada hubungan linier yang kuat.

```python
df['income_numeric'] = df['income_category'].map({'Low': 1, 'Medium': 2, 'High': 3})
correlation_matrix = df[['customer_age', 'income_numeric']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
